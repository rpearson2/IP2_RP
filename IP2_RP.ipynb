3.	Create an 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.
- a.	Create a plot using plt.subplots() and plot X versus X2
- b.	Create a scatter plot of X versus the exponential of X (np.exp(X))


```python
x= np.arange(0,100,1)
x_2 = x**2

fig, ax = plt.subplots()
ax.plot(x, x_2)

ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('Plot of X versus X^2')
plt.show()
```

```python
plt.scatter(x, np.exp(x))
plt.xlabel("x")
plt.ylabel("exponential of x")
plt.title("X vs exponential of X")
plt.show()
```

5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(int64) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
car_sales['Price']=car_sales['Price'].str.replace("$", "")
```

```python
car_sales.head()
```

```python
car_sales['Price']=car_sales['Price'].str.replace(",", "")
```

```python
car_sales.head()
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
car_sales.info()
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales.head()
```

12.	Create a subplot of multiple plots ( step 8,9,10,11).

```python
fig=plt.figure(figsize=[16,12])
fig.suptitle("Graph Comparison")
#figure 1
ax1=fig.add_subplot(221) 
ax1.set_title("Price vs Total Sales Line Plot")
ax1.plot(car_sales['Price'], car_sales['Total Sales'])
ax1.set_xlabel("Price")
ax1.set_ylabel("Total Sales")

# figure 2
ax2=fig.add_subplot(222)
ax2.set_title("Odometer (KM) vs Price Scatter Plot")
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax2.set_xlabel("Odometer (KM)")
ax2.set_ylabel("Price")

# figure 3
ax3=fig.add_subplot(223)
ax3.set_title("Make vs Odometer (KM) Bar Plot")
ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])
ax3.set_xlabel("Make")
ax3.set_ylabel("Odometer (KM)")

# figure 4
ax4=fig.add_subplot(224)
ax4.set_title("Histogram of Car Make")
ax4.hist(car_sales['Make'])
ax4.set_xlabel('Make')

plt.show()
```

16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
corr=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr()
print(corr)

sns.heatmap(corr, annot=True)
plt.show()
```

Price and Total Sales has very a positive, very weak to no correlation between the two variables. Price doesn't really affect total sales. 

Total Sales and Odometer (KM) has a weak, negative correlation between total sales and odometer (KM). So as Total sales goes up the odometer (KM) goes down. 

Price and Odometer (KM) has a moderately strong, negative correlation between price and odometer (km). So it can be assumed that as price goes up the value of odometer (KM) goes down.
